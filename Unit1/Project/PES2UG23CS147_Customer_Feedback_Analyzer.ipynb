```python
# Giving a few real reviews from google to see how it labels them.
while True:
    user_review = input("Enter a customer review (or type 'exit' to quit): ")
    if user_review.lower() == 'exit':
        break
    result = sentiment_analyzer(user_review)[0]
    label = result["label"]
    score = result["score"]

    print(f"Sentiment: {label} (Confidence: {score:.2f})")
    print("-" * 50)
```

We can observe that simple reviews have very accurate sentiment labels, but ambiguous reviews do not. This is a limitation of the model, since it is binary, where neutral reviews that can be debatable on being positive or negative is hard to label and unsurprisingly, the model does not do a good job. A surprising observation however is the confidence score. The model gives almost all reviews a confidence score of 1, even for "neutral" reviews like "I'm neutral about this product" was labelled as "NEGATIVE" with a score of 1.
